```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data1 = np.random.randint(0,101,40)
data2 = np.random.randint(0,101,40)
data3 = np.random.randint(0,101,40)

fig, axs = plt.subplots(2,1)

axs[0].plot(range(40), data1, color = 'orange', linewidth = 10, label = 'data 1')
axs[0].plot(range(40), data2, color = 'red', linestyle = '--', label = 'data 2')
axs[0].set_title('Line Plots')
axs[0].set_xlabel('Position in List')
axs[0].set_ylabel('Numeric Values')
axs[0].legend()

axs[1].scatter(range(40), data3, color = 'magenta', marker = 'D', label = 'data 3')
axs[1].set_title('Scatter Plot')
axs[1].set_xlabel('Position in List')
axs[1].set_ylabel('Numeric Value')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 16.02
b = 0.1124
R = 0.083144
n = 1

P, V = np.meshgrid(np.linspace(1, 10, 100), np.linspace(10, 30, 100))

T = (P + a * n ** 2 / V ** 2) * (V - n * b) / (n * R)

plt.pcolormesh(P, V, T, cmap = 'inferno', vmin = np.min(T), vmax = np.max(T))
plt.colorbar(label = 'Temperature')
plt.xlabel('Pressure')
plt.ylabel('Volume')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculating_pi(N):
    points = np.random.rand(N, 2)
    distance = np.linalg.norm(points, axis = 1)
    circle_points = np.sum(distance <= 1)
    calculated_pi = 4 * circle_points / N
    return calculated_pi

Ns = [int(1e3), int(1e4), int(1e5), int(1e6)]
pi_values = [calculating_pi(N) for N in Ns]
for N, pi in zip(Ns, pi_values):
    print(f"π approximation for N = {N}: {pi}")
    
N = int(1e4)
points = np.random.rand(N, 2)
distance = np.linalg.norm(points, axis = 1)
circle_points = points[distance <= 1]
outside_points = points[distance > 1]

plt.scatter(circle_points[:, 0], circle_points[:, 1], color = 'blue', label = 'Inside the Circle')
plt.scatter(outside_points[:, 0], outside_points[:, 1], color = 'red', label = 'Outside the Circle')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```
